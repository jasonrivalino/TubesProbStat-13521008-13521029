<h3> Penjelasan Informasi dalam Histogram dan Boxplot </h3>

1. Penjelasan Histogram: <br>
Data vertikal adalah frekuensi data, sedangkan data horizontal adalah range nilai dari sebuah data.
<br><br>
2. Penjelasan Boxplot: <br>
Box dalam sebuah visualisasi Boxplot menunjukkan kuartil 1, 2, dan 3 dari sebuah data. <br>
Garis tengah dalam box adalah nilai median dari sebuah data. <br>
Garis di bawah box adalah "minimum" dari sebuah data dengan rumus Q1 - 1.5*IQR. <br>
Garis di atas box adalah "maksimum" dari sebuah data dengan rumus Q3 + 1.5*IQR. <br>
Sedangkan, titik-titik di luar box adalah "outlier" yang merupakan data yang berada di luar range minimum dan maksimum. <br>

<h3> Visualisasi Histogram dan Boxplot </h3>

````python
import matplotlib.pyplot as plt

columns = df.columns

# histogram
for column in columns:
    plt.figure()
    plt.hist(df[column])
    plt.title(f'Histogram {column}')
    plt.xlabel(column)
    plt.ylabel('Freq')
    plt.show()
    print("\033[1m" + "Keterangan histogram:" + "\033[0m")
    print("Nilai data terkecil berada di: ", df[column].min())
    print("Nilai data dengan frekuensi tertinggi berada di: ", df[column].mode().iloc[0])
    print("Nilai data terbesar berada di: ", df[column].max())

    plt.figure()
    plt.boxplot(df[column], vert=False)
    plt.title(f'Boxplot {column}')
    plt.ylabel(column)
    plt.show()
    print("\033[1m" + "Keterangan boxplot:" + "\033[0m")
    print("Nilai median berada di: ", df[column].median())
    print("Nilai Q1 berada di: ", df[column].quantile(0.25))
    print("Nilai Q3 berada di: ", df[column].quantile(0.75))
    print("Nilai IQR berada di: ", df[column].quantile(0.75) - df[column].quantile(0.25))
    print("Nilai \"minimum\" berada di: ", df[column].quantile(0.25) - 1.5* abs((df[column].quantile(0.75) - df[column].quantile(0.25))))
    print("Nilai \"maksimum\" berada di: ", df[column].quantile(0.75) + 1.5*abs((df[column].quantile(0.75) - df[column].quantile(0.25))))
    print()
````

<h2> 3. Menentukan setiap kolom numerik berdistribusi normal atau tidak. </h2><br>
Menentukan setiap kolom numerik berdistribusi normal atau tidak dengan menggunakan normality test yang dikaitkan dengan histogram plot.

<h3> Penjelasan cara yang dipakai </h3>

Untuk penerapan cara pengeceken normality test, kami menggunakan cara <b>D'Agostino-Pearson K-squared test</b>, cara yang dipakai ini bergantung pada nilai hasil test skewness dan kurtosis yang kemudian digabungkan dengan rumus berikut:

$$Z_s^2 + Z_k^2 \sim X^2 (2)$$

Keterangan: <br>
$Z_s^2$ = z-score dari test skewness <br>
$Z_k^2$ = z-score dari test kurtosis <br>
$X^2 (2)$ = aproksimasi chi-squared terdistribusi yang memiliki derajat kebebasan 2 <br>

Untuk penerapannya, dilakukan dengan cara mencari nilai p-value menggunakan fungsi `normaltest` yang ada pada library `scipy`. Setelah nilai p-value didapatkan, kemudian dibandingkan dengan nilai tetap alpha yaitu 0.05. Kondisi perbandingannya adalah sebagai berikut:
1. Jika nilai p-value lebih besar daripada nilai alpha, data memenuhi terdistribusi normal
2. Jika nilai p-value lebih kecil daripada nilai alpha, data tidak memenuhi terdistribusi normal

Untuk membantu pembuktian, terdapat juga visualisasi histogram dengan menggunakan library `matplotlib`. Data yang terdistribusi normal akan memiliki bentuk kurva yang menyerupai bentuk lonceng.

<h3> Pengecekan distribus normal pada setiap kolom beserta visualisasi histogramnya </h3>

````python
from scipy import stats
import matplotlib.pyplot as plt

for column in columns:
    data = df[column]
    stat, pVal = stats.normaltest(data)
    alpha = 0.05 # nilai alpha tetap
    
    # histogram
    plt.figure()
    count, bins, ignored = plt.hist(data, 50, density=True)

    # tambah kurva lonceng
    mean, std = data.mean(), data.std()
    x = np.linspace(data.min(), data.max(), 100)
    y = stats.norm.pdf(x, mean, std)
    plt.plot(x, y * count.max() / y.max())

    # tampil plot
    plt.title(f'Histogram {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
    
    print("Nilai p-value pada kolom",column,"adalah:",pVal)
    print("")
    print("Hasil:")
    if pVal > alpha:
        print("Data",column,"berdistribusi normal karena p-value lebih besar dari alpha (",pVal,">",alpha,")")
    else:
        print("Data",column,"TIDAK berdistribusi normal karena p-value lebih kecil dari alpha (",pVal,"<",alpha,")")
````

<h2>Melakukan test hipotesis 1 sampel,</h2>

a. Nilai rata-rata pH di atas 3.29?

````python
# Soal 4.1 

import scipy.stats as stats 
import math

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 3.29

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 > 3.29 (One Tailed - Right Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05
zAlpha = abs(stats.t.ppf(0.05, 999)) # dimutlak karena Right Tailed

# Step 4: hitung uji statistik dengan rumus z
mean = df['pH'].mean()
std = df['pH'].std()
z = (mean - h0) / (std/math.sqrt(1000))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Right Tailed:")
print("t >", zAlpha)

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z < zAlpha):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",z,"<",zAlpha,")")
if (z > zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",zAlpha,")")
    
print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = stats.t.sf(abs(z), 999)
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
if (pVal > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal,">",alpha,")")
if (pVal <= alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal,"<",alpha,")")
````

b. Nilai rata-rata Residual Sugar tidak sama dengan 2.50?

````python
# Soal 4.2

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 2.50

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 != 2.50 (Two Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05/2
zAlpha = stats.t.ppf(0.025, 999)

# Step 4: hitung uji statistik dengan rumus z
mean = df['residual sugar'].mean()
std = df['residual sugar'].std()
z = (mean - h0) / (std/math.sqrt(1000))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Two Tailed:")
print("t <", zAlpha, "dan t >", abs(zAlpha))

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z > zAlpha and z < abs(zAlpha)):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",zAlpha,"<",z,"<",abs(zAlpha),")")
if (z < zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,"<",zAlpha,")")
if (z > abs(zAlpha)):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",abs(zAlpha),")")

print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = 2*(stats.t.sf(abs(z), 999))
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
# pVal dibagi dua karena hanya meninjau salah satu sisi saja
if (pVal/2 > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal/2,">",alpha,")")
if (pVal/2 < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal/2,"<",alpha,")")
````

c. Nilai rata-rata 150 baris pertama kolom sulphates bukan 0.65?

````python
# Soal 4.3

import scipy.stats as stats 

# Pre-Step = menentukan bagian baris tabel yang ingin ditinjau
sampleTable = df.loc[0:149]

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 0.65

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 != 0.65 (Two Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05/2
zAlpha = stats.t.ppf(0.025, 999)

# Step 4: hitung uji statistik dengan rumus z
mean = sampleTable['sulphates'].mean()
std = sampleTable['sulphates'].std()
z = (mean - h0) / (std/math.sqrt(150))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Two Tailed:")
print("t <", zAlpha, "dan t >", abs(zAlpha))

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z > zAlpha and z < abs(zAlpha)):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",zAlpha,"<",z,"<",abs(zAlpha),")")
if (z < zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,"<",zAlpha,")")
if (z > abs(zAlpha)):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",abs(zAlpha),")")
    
print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = 2*(stats.t.sf(abs(z), 149))
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
# pVal dibagi dua karena hanya meninjau salah satu sisi saja
if (pVal/2 > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal/2,">",alpha,")")
if (pVal/2 < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal/2,"<",alpha,")")
````

d. Nilai rata-rata total sulfur dioxide di bawah 35 ?

````python
# Soal 4.4

import scipy.stats as stats 

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 35

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 < 35 (One Tailed - Left Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05
zAlpha = stats.t.ppf(0.05, 999)

# Step 4: hitung uji statistik dengan rumus z
mean = df['total sulfur dioxide'].mean()
std = df['total sulfur dioxide'].std()
z = (mean - h0) / (std/math.sqrt(1000))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Left Tailed:")
print("t <", zAlpha)

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z > zAlpha):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",z,">",zAlpha,")")
if (z < zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,"<",zAlpha,")")

print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = stats.t.cdf(abs(z), 999)
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
if (pVal > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal,">",alpha,")")
if (pVal < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal,"<",alpha,")")
````

e. Proporsi nilai total Sulfat Dioxide yang lebih dari 40, adalah tidak sama dengan 50% ?

````python
# Soal 4.5

# Pre-Step = menentukan bagian baris tabel yang ingin ditinjau
sampleTable = df.loc[df['total sulfur dioxide']>40]
totalData = sampleTable.shape[0]/1000

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 0.5

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 != 0.5 (Two Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05/2
zAlpha = stats.t.ppf(0.025,(totalData*1000)-1)

z = (totalData - h0) / math.sqrt((0.5*0.5)/1000)

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Two Tailed:")
print("t <", zAlpha, "dan t >", abs(zAlpha))

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z > zAlpha and z < abs(zAlpha)):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",zAlpha,"<",z,"<",abs(zAlpha),")")
if (z < zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,"<",zAlpha,")")
if (z > abs(zAlpha)):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",abs(zAlpha),")")
    
print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = 2*(stats.t.sf(abs(z), (totalData*1000)-1))
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
# pVal dibagi dua karena hanya meninjau salah satu sisi saja
if (pVal/2 > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal/2,">",alpha,")")
if (pVal/2 < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal/2,"<",alpha,")")
````

5. Melakukan test hipotesis 2 sampel,

a. Data kolom fixed acidity dibagi 2 sama rata: bagian awal dan bagian akhir kolom.
Benarkah rata-rata kedua bagian tersebut sama?

````python
# Soal 5.1

# Pre-Step = bagi dua tabel menjadi sama
upperTable = df.loc[:499]
lowerTable = df.loc[500:]

# Step 1: Menentukan nilai hipotesis 0 (H0)
h0 = 0

# Step 2: Menentukan nilai hipotesis 1 (H1)
# mean H1 != 0 (Two Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05/2
zAlpha = stats.t.ppf(0.025, 998)

# Step 4: hitung uji statistik dengan rumus z
mean1 = upperTable['fixed acidity'].mean()
std1 = upperTable['fixed acidity'].std()
mean2 = lowerTable['fixed acidity'].mean()
std2 = lowerTable['fixed acidity'].std()

z = (((mean1-mean2)-h0)/(math.sqrt(pow(std1,2)/500+pow(std2,2)/500)))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Two Tailed:")
print("t <", zAlpha, "dan t >", abs(zAlpha))

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z > zAlpha and z < abs(zAlpha)):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",zAlpha,"<",z,"<",abs(zAlpha),")")
if (z < zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,"<",zAlpha,")")
if (z > abs(zAlpha)):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",abs(zAlpha),")")
    
print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = 2*(stats.t.sf(abs(z), 998))
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
# pVal dibagi dua karena hanya meninjau salah satu sisi saja
if (pVal/2 > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal/2,">",alpha,")")
if (pVal/2 < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal/2,"<",alpha,")")
````

e. Proporsi nilai setengah bagian awal alcohol yang lebih dari 7, adalah lebih besar
daripada, proporsi nilai yang sama di setengah bagian akhir alcohol?

````python
# Soal 5.5

# Pre-Step = menentukan bagian baris tabel yang ingin ditinjau
preUpperTable = df.loc[:499]
upperTable = preUpperTable.loc[df['alcohol']>7]
lowerTable = df.loc[500:]

totalDataUpper = upperTable.shape[0]
totalDataLower = lowerTable.shape[0]

# Step 1: Menentukan nilai hipotesis 0 (H0)
# h0 = (p1 = p2)

# Step 2: Menentukan nilai hipotesis 1 (H1)
# h0 = (p1 > p2) (One Tailed - Right Tailed)

# Step 3: inisiasi nilai alpha
alpha = 0.05
zAlpha = abs(stats.t.ppf(0.05, 998)) # dimutlak karena Right Tailed

# Step 4: hitung uji statistik dengan rumus z
p = (totalDataUpper + totalDataLower) / (1000)

z = (((totalDataLower/500) - (totalDataUpper/500)) / math.sqrt(p*(1-p)*(1/500+1/500)))

print("Nilai z adalah:")
print(z)

print("")
print("Nilai zAlpha adalah:")
print(zAlpha)

# Menentukan daerah kritis
print("")
print("Critical Region dari Right Tailed:")
print("t >", zAlpha)

print("")
print("Pengecekan pertama (tanpa p-val):")
if (z < zAlpha):
    print("Tes menerima H0 karena z tidak masuk Critical Region (",z,"<",zAlpha,")")
if (z > zAlpha):
    print("Tes menolak H0 karena z masuk Critical Region (",z,">",zAlpha,")")
    
print("")
print("Nilai alpha adalah:")
print(alpha)

# Step 5: Menghitung p-value sesuai dengan uji statistik
print("")
print("Nilai p-value:")
pVal = stats.t.sf(abs(z), 998)
print(pVal)

# Step 6: Menentukan penolakan atau penerimaan berdasarkan nilai pval dan alpha
print("")
print("Pengecekan kedua (dengan p-val):")
if (pVal > alpha):
    print("Tes menerima H0 karena p-val lebih besar dari alpha (",pVal,">",alpha,")")
if (pVal < alpha):
    print("Tes menolak H0 karena p-val lebih kecil dari alpha (",pVal,"<",alpha,")")
````
